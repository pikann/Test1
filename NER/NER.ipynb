```
import re
import os

list_path=['/content/VLSP2018-NER-Test-Domains/', '/content/VLSP2018-NER-dev/',
           '/content/VLSP2018-NER-train-Jan14/']

sentences=[]
labels=[]

for file_path in list_path:
    all_folder=os.listdir(file_path)

    for folder_name in all_folder:
        file_in_folder=os.listdir(file_path+folder_name)

        for file_name in file_in_folder:
            f = open(file_path+folder_name+'/'+file_name, "r")
            s=f.read()
            s1=s.split("\n")
            s3=[]
            for s2 in s1:
                if (s2!=""):
                    s4=s2.split(". ")
                    for s5 in s4:
                        if (s5!=""):
                            s3.append(s5)
            s4=[]
            for s2 in s3:
                s4.append(re.findall(r"[\w']+", s2))

            for s2 in s4:
                if (len(s2)>500):
                    continue
                sentence=[]
                label=[]
                i=0
                while i<len(s2):
                    if (s2[i]=='ENAMEX'):
                        i+=2
                        if (i+1>=len(s2)):
                            break
                        if (s2[i]=='PERSON' or s2[i]=='LOCATION'):
                            i+=1
                            sentence.append(s2[i].lower())
                            label.append(1)
                            i+=1
                            if (i>=len(s2)):
                                break
                            while (s2[i]!='ENAMEX'):
                                sentence.append(s2[i].lower())
                                label.append(2)
                                if (i+1<len(s2)):
                                    i+=1
                                else:
                                    break
                        else:
                            i+=1
                            while (s2[i]!='ENAMEX'):
                                sentence.append(s2[i].lower())
                                label.append(0)
                                i+=1
                                if (i>=len(s2)):
                                    break
                            i+=1
                    else:
                        sentence.append(s2[i].lower())
                        label.append(0)
                    i+=1
                sentences.append(sentence)
                labels.append(label)
```

```
import pandas as pd
list_word=[]
for sentence in sentences:
    list_word+=sentence
list_word=pd.DataFrame(list_word)
print(list_word)
```

```
words = list(list_word[0].unique())
print(words)
```

```
word2idx = {w : i + 2 for i, w in enumerate(words)}
word2idx["UNK"] = 1
word2idx["PAD"] = 0
print(word2idx)
```

```
idx2word = {i: w for w, i in word2idx.items()}
```

```
max_len=0
for sentence in sentences:
    if (len(sentence)>max_len):
        max_len=len(sentence)
print(max_len)
```

```
from tensorflow.keras.preprocessing.sequence import pad_sequences

X = [[word2idx[w] for w in s] for s in sentences]
X = pad_sequences(maxlen = max_len, sequences = X, padding = "post", value = word2idx["PAD"])
y = pad_sequences(maxlen = max_len, sequences = labels, padding = "post", value = 3)
```

```
print(y)
```

```
from tensorflow.keras.utils import to_categorical
y = [to_categorical(i, num_classes = 4) for i in y]
print(y)
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Bidirectional, LSTM, TimeDistributed, Dense, Dropout, BatchNormalization
from tf2crf import CRF

model=Sequential()
model.add(Embedding(input_dim=len(word2idx), output_dim=40, input_length=max_len, mask_zero=False))
model.add(Bidirectional(LSTM(units=100, return_sequences=True, recurrent_dropout=0.6, dropout=0.6)))
model.add(TimeDistributed(Dense(100, activation="relu")))
model.add(Dropout(0.6))
crf = CRF(4)
model.add(crf)

model.compile(optimizer="adam", loss=crf.loss, metrics=[crf.accuracy])
```

```
model.summary()
```

```
model.load_weights("/content/drive/My Drive/NER/model.h5")
```

```
from tensorflow.keras.callbacks import ModelCheckpoint
checkpoint = ModelCheckpoint(filepath = '/content/drive/My Drive/NER/model.h5',
                        verbose = 0,
                        mode = 'auto',
                        save_best_only = True,
                        monitor='val_loss')

model.fit(np.array(X_train), np.array(y_train), batch_size=128, epochs=50, 
          validation_data=(np.array(X_test), np.array(y_test)), callbacks=[checkpoint])
```
